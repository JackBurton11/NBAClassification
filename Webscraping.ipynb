```python
#writing loop to scrape 4 years of eastern conference team info into 1 dataframe

years = np.arange(2015, 2019, 1)

base = 'https://www.basketball-reference.com/leagues/NBA_{}_standings.html'

team_wl = pd.DataFrame()

for year in years:
    url = base.format(year)
    req = requests.get(url)
    soup = BeautifulSoup(req.content, 'html.parser')
    table1 = soup.find(name='table', attrs={'id':'divs_standings_E'})
    html_str1 = str(table1)
    df = pd.read_html(html_str1)[0]
    df['Year'] = year
    
    team_wl = team_wl.append(df)
```

```python
#writing loop to scrape 4 years of western conference division info into 1 dataframe

years = np.arange(2015, 2019, 1)

base = 'https://www.basketball-reference.com/leagues/NBA_{}_standings.html'

team_wl1 = pd.DataFrame()

for year in years:
    url = base.format(year)
    req = requests.get(url)
    soup = BeautifulSoup(req.content, 'html.parser')
    table1 = soup.find(name='table', attrs={'id':'divs_standings_W'})
    html_str1 = str(table1)
    df = pd.read_html(html_str1)[0]
    df['Year'] = year
    
    team_wl1 = team_wl1.append(df)
```
